```python
df=df.convert_objects(convert_numeric=True)
drop_out=df.isnull().sum()
drop_out_rate=drop_out/df.count()
```

```python
# fill the empty places
df=df.fillna(method="backfill")
```

```python
# extracting the mains supply meter readings
agg_elec_main=df[['Device 1---MOD-EL01-10-65-160-155-ME1','Device 1---MOD-EL02-10-65-160-155-ME1','Device 1---MOD-EL03-10-65-160-155-ME1']]
```

```python
agg_elec_main=agg_elec_main.convert_objects(convert_numeric=True)
agg_elec_main=agg_elec_main.fillna(method="backfill")
```

```python
elec_main=agg_elec_main.diff(periods=1)

elec_main['energy']=elec_main.sum(axis=1)
```

```python
elec_main[abs(elec_main.values)>40000]=np.nan
elec_main[elec_main.values<0]=np.nan
```

```python
daily_elec_main=elec_main.groupby(elec_main.index.date).sum()
daily_elec_main.plot(kind='bar',figsize=(20,6))
plt.xlabel('Days', fontsize= 15)
plt.ylabel('kWh', fontsize= 15)
plt.title("Daily electricity consumption")
```

```python
electricty_consumption_day = daily_elec_main["energy"].mean(axis=0, skipna=True)
print electricty_consumption_day, "kWh"
```

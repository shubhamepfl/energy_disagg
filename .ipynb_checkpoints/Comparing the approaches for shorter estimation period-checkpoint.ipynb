# Contextual disaggregation method 

```python
# ref_x=pd.DataFrame(columns=[states_table_ds["refrigerator1"], states_table_ds["refrigerator2"]], index=states_table_ds.index)
ref=states_table_ds.loc[:,"refrigerator1":"refrigerator2"]
light=states_table_ds.loc[:,"lighting1":"lighting2"]
dish=states_table_ds.loc[:,"dishwaser1":"dishwaser2"]
y_tot=mains_ds["Mains"]
```

```python
from cvxpy import *
import numpy as np

# optimisation
i=states_table_ds.index[0]
ref_dis = pd.DataFrame(data=np.nan, index=states_table_ds.index, columns=["refrigerator"])
light_dis = pd.DataFrame(data=np.nan, index=states_table_ds.index, columns=["lighting"])
dish_dis = pd.DataFrame(data=np.nan, index=states_table_ds.index, columns=["dishwaser"])
while i <= (states_table_ds.tail(n=1).index):
    start_time=i
    print start_time
    end_time=i+timedelta(hours=t_est)
    ref_x=ref[start_time:end_time]
    light_x=light[start_time:end_time]
    dish_x=dish[start_time:end_time]
    y=y_tot[start_time:end_time]
    n= y.shape[0]
    ref_pred = Variable(n)
    light_pred= Variable(n)
    dish_pred= Variable(n)
    theta_ref= Variable(2)
    theta_light= Variable(2)
    theta_dish= Variable(2)
    objective = Minimize(sum_entries(square(ref_pred-ref_x.values*theta_ref)+square(light_pred-light_x.values*theta_light)+square(dish_pred-dish_x.values*theta_dish)))
    # constraints = [ref_pred==ref_x.values*theta_ref, light_pred==light_x.values*theta_light, ref_pred + light_pred==y.values, light_pred>=0, ref_pred>=0]
    constraints = [ref_pred + light_pred + dish_pred ==y.values, light_pred>=0, ref_pred>=0, dish_pred>=0]
    prob = Problem(objective, constraints)
    prob.solve()
    ref_dis.loc[start_time:end_time]=ref_pred.value
    light_dis.loc[start_time:end_time]=light_pred.value
    dish_dis.loc[start_time:end_time]=dish_pred.value
    i+=timedelta(hours=t_est)
        
```

What is happening here is that because we have used only on/off states as context variables; for the dishwasher which only comes on at very specific instances 
Now when the dishwasher comes on, dish_pred cannot take all the value as its cost will blow up, so to respect the constrains of total power this extra power is distributed amaong different appliances. So we see spikes in the power of other appliances as well

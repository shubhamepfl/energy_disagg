```python
df_office.head()
```

```python
from sklearn.cluster import KMeans
def clustering(data, n_clusters):
    kmeans= KMeans(init='k-means++', n_clusters=n_clusters)
    kmeans.fit(data)
    return kmeans.cluster_centers_, kmeans.labels_
```

```python
cluster, labels=clustering(df_office, 3)

for i in xrange(5):
    print df_sub["Device 1---MOD-WMZ07-10-65-160-164-ME1"][labels==i].mean()
```

```python
from sklearn import decomposition
pca= decomposition.PCA(n_components=2, whiten=True)
pca.fit(df_office)
X_office= pca.transform(df_office)
plt.scatter(X_office[1:,0],X_office[1:,1], c=labels_office)
plt.legend()
```
